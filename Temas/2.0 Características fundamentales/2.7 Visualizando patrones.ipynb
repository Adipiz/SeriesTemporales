# Visualizando patrones en series temporales

```python
# paqueterias

import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
# Cargar los datos: Datos corresponde a la toma de temperaturas de un edificio acda 30 min.

dfTi = pd.read_csv("../../DataSets/Ti_data.csv", index_col = 0, parse_dates = True)
dfTi
```

```python
# VisualizaciÃ³n de datos

dfTi.plot(figsize = (21,8))
plt.title("Internal Temperature (Ti)", size = 17)
plt.ylabel("ÂºC")
plt.show()
```

```python
# Visualizando patrones diurnos
# Separando por dÃ­a y hora

dfTi['date'] = dfTi.index.normalize()
dfTi['time'] = dfTi.index.time.astype(str)
dfTi
```

```python
# Vamos a crear un Df a partir del nuevo.

dfPivot = dfTi.pivot(index = 'date', columns = 'time', values = 'Ti')
dfPivot
```

```python
# Grafica
dfPivot.T.plot(legend = False, figsize = (20,8), color = 'C0', alpha = 0.1)
plt.title("Internal Temperature $Ti$", size = 21 )
plt.ylabel("Ti (ÂºC)")
plt.show()
```

```python
# Cambiando el horario de 00:00 a 18:00
# son 36 medias horas

dfRoll = pd.DataFrame(index = dfPivot.index,
                     columns = np.roll(dfPivot.columns, -36),
                     data = np.roll(dfPivot.values, -36))
dfRoll
```

```python
dfRoll.T.plot(legend = False, figsize = (20,8), color = 'C0', alpha = 0.1)
plt.show()
```

```python
day_colors = { 'Monday': 'C0', 
               'Tuesday': 'C1',
               'Wednesday': 'C1',
               'Thursday': 'C1',
               'Friday': 'C1',
               'Saturday': 'C2',
               'Sunday': 'C3'}

# Plot
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

for day_name, color in day_colors.items():
    # NOTE: we are plotting the transposed DataFrame
    dfRoll[dfRoll.index.day_name() == day_name].T.plot(
        ax=ax, color=color, alpha=0.1, legend=None)

ax.set_xlim([0, 47])
ax.set_ylim(17.75, 22)

#LÃ­neas verticales definiendo los horarios de calefacciÃ³n
ax.axvline(30, alpha=0.25, color='black', linestyle='--')
ax.axvline(40, alpha=0.25, color='black', linestyle='--')
ax.axvline(42, alpha=0.25, color='black', linestyle='--')

ax.set_ylabel('$ËšC$')
ax.set_title('Internal Temperature $(T_i)$');

# Leyenda:
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='C0', lw=2),
                Line2D([0], [0], color='C1', lw=2),
                Line2D([0], [0], color='C2', lw=2),
                Line2D([0], [0], color='C3', lw=2)]

ax.legend(custom_lines, ['Mon', 'Tue-Fri', 'Sat', 'Sun']);
    
```
