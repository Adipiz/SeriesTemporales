# SARIMA (p,d,q)(P,D,Q,s)

## Seasonal Autoregressive Integrated Moving Averages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# tools

from statsmodels.tsa.statespace.sarimax import SARIMAX

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf # (p,q) orders
from statsmodels.tsa.seasonal import seasonal_decompose       # ETS plots


# Warnings 
import warnings
warnings.filterwarnings("ignore")

# Dataset
df = pd.read_csv("../../DataSets/co2_mm_mlo.csv")
```

```python
df.head()
```

```python
# Añadir una variable "date"
df['date'] = pd.to_datetime(dict(year=df['year'], month = df['month'], day = 1))
```

```python
# Hacer que la variable "data" sea el índice

df.set_index('date', inplace = True)
df.index.freq = 'MS'
df.head()
```

```python
# plotting data

title  = "Monthly mean CO2 levels (ppm) over Mauna Loa, Hawaii"
ylabel = "parts per million"
xlabel = ''

df.interpolated.plot(figsize = (12,6))
plt.title(title)
plt.ylabel(ylabel)
plt.show()
```

## Descomposición en tres factores
   - Tendencia, estacional, residual

```python
results = seasonal_decompose(df['interpolated'], model = 'add')
results.plot()
```

### División de datos

```python
train = df.iloc[:717]
test = df.iloc[717:]
```

## Modelo SARIMA(0,1,1)(1,0,1,12)

```python
model = SARIMAX(train['interpolated'], order = (0,1,1), seasonal_order = (1,0,1,12))
results = model.fit()
results.summary()
```

### Obtaining predicted values



```python
start = len(train)
end = len(train) + len(test) - 1

predictions = results.predict(start = start, end = end).rename('SARIMA(0,1,1)(1,0,1,12) Predictions')
```

```python
# Comparing predictions to expected values

for i in range(len(predictions)):
    print(f"predicted = {predictions[i]}, expected={test['interpolated'][i]}")
```

```python
title  = "Monthly mean CO2 levels (ppm) over Mauna Loa, Hawaii"
ylabel = "parts per million"
xlabel = ''

test.interpolated.plot(legend = True,figsize = (12,6))
predictions.plot(legend = True)
plt.title(title)
plt.ylabel(ylabel)
plt.show()
```

AL parecer el modelo se ajusta bien a los datosm de prueba. Por tano, si convence, podemos reentrenar el modelo con toda la data para hacer predicciones.

## Reentrenando el MODELO con TODOS los datos

```python
model = SARIMAX(df['interpolated'], order = (0,1,1), seasonal_order = (1,0,1,12))
results = model.fit()
fcast = results.predict(len(df), len(df) + 36).rename('SARIMA(0,1,1)(1,0,1,12) Forecast')
```

## Graficando predicciones

```python
title  = "Monthly mean CO2 levels (ppm) over Mauna Loa, Hawaii"
ylabel = "parts per million"
xlabel = ''

df.interpolated.plot(legend = True,figsize = (12,6))
plt.title(title)
plt.ylabel(ylabel)
plt.xlabel(xlabel)
fcast.plot(legend = True)
plt.show()
```
